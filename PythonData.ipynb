```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
enrollment = pd.read_csv("~/Desktop/chia2018/2017/enrollment_premiums/enrollment_by_employer_size.csv", thousands=",")
#Reading the dataset in a dataframe using Pandas
num = 7
enrollment.head(num) #prints <num> rows
```

```python
months = enrollment[enrollment.columns[1:5]]
mth_cols = ['months_2014','months_2015','months_2016']
months
```

```python
premiums = pd.read_csv("~/Desktop/chia2018/2017/enrollment_premiums/premiums_by_employer_size.csv", thousands=",")
premiums[pre_cols] = premiums[pre_cols].replace({'\$': '', ',': ''}, regex=True)
premiums[pre_cols] = premiums[pre_cols].apply(pd.to_numeric)

num = 7

agg_premiums = premiums[premiums.columns[1:5]]
agg_cols = ['agg_premiums_2014','agg_premiums_2015','agg_premiums_2016']
```

```python
new_df = pd.DataFrame(columns = ['Employer Size','pmpm_2014','pmpm_2015','pmpm_2016'])
new_df['Employer Size'] = enrollment['Employer Size']
new_df['pmpm_2014'] = agg_premiums['agg_premiums_2014']/months['months_2014']
new_df['pmpm_2015'] = agg_premiums['agg_premiums_2015']/months['months_2015']
new_df['pmpm_2016'] = agg_premiums['agg_premiums_2016']/months['months_2016']
new_df
```

```python
pmpm = ['pmpm_2014','pmpm_2015','pmpm_2016']
loop_df = pd.DataFrame(columns = ['Employer Size'] + pmpm)
for i in range(3):
    agg = agg_cols[i]
    mth = mth_cols[i]
    loop_df[pmpm[i]] = agg_premiums[agg]/months[mth]
loop_df['Employer Size'] = enrollment['Employer Size']

loop_df
```
